```python
df1 = pd.read_csv('/content/drive/MyDrive/2024BIGCON/최종_유동인구수_데이터셋.csv')
df2 = pd.read_csv('/content/drive/MyDrive/2024BIGCON/버스정류장현황_20240331.csv')
df3 = pd.read_csv('/content/drive/MyDrive/2024BIGCON/출발지읍면동별_관광객방문수.csv')
df4 = pd.read_csv('/content/drive/MyDrive/2024BIGCON/읍면동별_인구수.csv')
```

```python
df2.rename(columns={'인원': '승하차수'}, inplace=True)
```

```python
# 특정 위치의 값을 추출하는 함수 (세 번째 위치에 있는 값을 가져옴)
def extract_specific_area(address):
    parts = address.split()
    if len(parts) >= 3:
        return parts[2]
    return None

# 새로운 변수로 추출한 값을 저장
df2['읍면동명'] = df2['지번주소'].apply(extract_specific_area)
```

```python
df1['읍면동명'] = df1['읍면동명'].replace('관설동', '반곡관설동')
df2['읍면동명'] = df2['읍면동명'].replace('반곡동', '반곡관설동')
df2['읍면동명'] = df2['읍면동명'].replace('관설동', '반곡관설동')
df3['출발지_읍면동명'] = df3['출발지_읍면동명'].replace('명륜2동', '명륜동')
df3.rename(columns={'출발지_읍면동명': '읍면동명'}, inplace=True)
```

```python
# 두 데이터프레임을 '읍면동명'을 기준으로 병합
df2 = df2.merge(df1[['읍면동명', '총_유동인구수']], on='읍면동명', how='left')
# 두 데이터프레임을 '읍면동명'을 기준으로 병합
df2 = df2.merge(df4[['읍면동명', '인구수']], on='읍면동명', how='left')
# 두 데이터프레임을 '읍면동명'을 기준으로 병합
df2 = df2.merge(df3[['읍면동명', '관광객방문수']], on='읍면동명', how='left')
```
